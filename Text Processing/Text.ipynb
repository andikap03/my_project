```python
import pandas as pd
import numpy as np

DATA = pd.read_csv("/content/Data/JAM_Reviews_En.csv")
REVIEW_DATA = DATA.dropna()
REVIEW_DATA['Review'] = REVIEW_DATA['Review_En']
REVIEW_DATA['Review'].head()
```

```python
REVIEW_DATA['Review'] = REVIEW_DATA['Review'].str.lower()

print('Case Folding Result :\n')
print(REVIEW_DATA['Review'].head(5))
```

```python
import string 
import re
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

def remove_special(text):
    text = text.replace('\\t',' ').replace('\\n'," ").replace('\\u'," ").replace('\\'," ")
    text = text.encode('ascii','replace').decode('ascii')
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)|(\(\W+\))"," ", text).split())
    text = ' '.join(re.sub(r"\d+","",text).split())
    text = ' '.join(re.sub('\s+',' ',text).split())
    text = text.translate(str.maketrans("","",string.punctuation))
    text = text.strip()
    return text.replace("http://"," ").replace("https://", " ")

REVIEW_DATA['Review'] = REVIEW_DATA['Review'].apply(remove_special)
REVIEW_DATA['Review']
```

```python
import nltk
nltk.download('punkt')
```

```python
def word_tokenize_wrapper(text):
  return word_tokenize(text)

REVIEW_DATA['Review_Tokens'] = REVIEW_DATA['Review'].apply(word_tokenize_wrapper)
print('Tokenizing Result : \n')
REVIEW_DATA['Review_Tokens'].head(5)
```

```python
def freqDist_wrapper(text):
  return FreqDist(text)

REVIEW_DATA['review_tokens_fdist'] = REVIEW_DATA['Review_Tokens'].apply(freqDist_wrapper)
print('Frequency Tokens : \n')
print(REVIEW_DATA['review_tokens_fdist'].apply(lambda x : x.most_common()))
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
list_stopwords = stopwords.words('english')
len(list_stopwords)
```

```python
list_stopwords = set(list_stopwords)

def stopwords_removal(words):
  return [word for word in words if word not in list_stopwords]

REVIEW_DATA['Review_tokens_WSW'] = REVIEW_DATA['Review_Tokens'].apply(stopwords_removal)
REVIEW_DATA['Review_tokens_WSW']
```

```python
REVIEW_DATA['Review_tokens_WSW_fdist'] = REVIEW_DATA['Review_tokens_WSW'].apply(freqDist_wrapper)
print(REVIEW_DATA['Review_tokens_WSW_fdist'].apply(lambda x : x.most_common()))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=1000)
vectors = vectorizer.fit_transform(REVIEW_DATA['Review'])

words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
words_df.head()
```

```python
x = words_df
y = REVIEW_DATA.Gender_En
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB

logreg = LogisticRegression(C=1e9, solver='lbfgs', max_iter=1000)
logreg.fit(x, y)
forest = RandomForestClassifier(n_estimators=50)
forest.fit(x,y)
svc = LinearSVC()
svc.fit(x,y)
bayes = MultinomialNB()
bayes.fit(x,y)
```

```python
pd.set_option("display.max_colwidth", 200)

unknown = pd.DataFrame({'content':["I love love love love this kitten",
                                   "I good good good good this keyboard",
                                   "I'm not sure how I feel about toast",
                                   " i feel that it was sold before"]})
unknown
```

```python
unknown_vectors = vectorizer.transform(unknown.content)
unknown_words_df = pd.DataFrame(unknown_vectors.toarray(), columns=vectorizer.get_feature_names())
unknown_words_df.head()
```

```python
# Prediksi dengan model classifier

# Logistic Regression prediction dan probabilities
unknown['pred_logreg'] = logreg.predict(unknown_words_df)
unknown['pred_logreg_proba'] = logreg.predict_proba(unknown_words_df)[:,1]

# Random Forest prediction dan probabilities
unknown['pred_forest'] = forest.predict(unknown_words_df)
unknown['pred_forest_proba'] = forest.predict_proba(unknown_words_df)[:,1]

# SVC prediction
unknown['pred_svc'] = svc.predict(unknown_words_df)

# Bayes prediction dan probabilities
unknown['pred_bayes'] = bayes.predict(unknown_words_df)
unknown['pred_bayes_probs'] = bayes.predict_proba(unknown_words_df)[:,1]

unknown
```
