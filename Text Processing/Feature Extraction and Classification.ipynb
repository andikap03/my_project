```python
import pandas as pd

df = pd.read_csv("/content/sentiment140-subset.csv", nrows=30000) #nrows untuk membatasi jumlah kolom dataset
df.head()
```

```python
# Mencari ukuran data
df.shape
```

```python
# Mencari jumlah distribusi data
df.polarity.value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df.text
```

```python
vectorizer = TfidfVectorizer(max_features=1000)
#Tf = term frequency
#idf = log(jumlah data/jumlah data keseluruhan) = untuk mencari pembeda
#Tfidf = merumuskan berat (0-1) apakah kata tersebut merupakan pembeda yg baik
 
vectors = vectorizer.fit_transform(df.text)

words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
words_df.head()
```

```python
x = words_df
y = df.polarity
```

**KLASIFIKASI**

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB
```

```python
%%time
# Buat dan train Logistic Regression
logreg = LogisticRegression(C=1e9, solver='lbfgs', max_iter=1000)
logreg.fit(x, y)
```

```python
%%time
# Buat dan train Random Forest Classifier
forest = RandomForestClassifier(n_estimators=50)
forest.fit(x,y)
```

```python
%%time
# Buat dan train Linear Support Vector Classifier
svc = LinearSVC()
svc.fit(x,y)
```

```python
%%time
# Buat dan train Multinomial Naive Bayes Classifier
bayes = MultinomialNB()
bayes.fit(x,y)
```

```python
# Buat Test Data
pd.set_option("display.max_colwidth", 200)

unknown = pd.DataFrame({'content':["I love love love love this kitten",
                                   "I hate hate hate hate this keyboard",
                                   "I'm not sure how I feel about toast",
                                   "Did you see the baseball game yesterday?"]})
unknown
```

```python
# Vectorizing Data Test
unknown_vectors = vectorizer.transform(unknown.content)
unknown_words_df = pd.DataFrame(unknown_vectors.toarray(), columns=vectorizer.get_feature_names())
unknown_words_df.head()
```

```python
# Prediksi dengan model classifier

# Logistic Regression prediction dan probabilities
unknown['pred_logreg'] = logreg.predict(unknown_words_df)
unknown['pred_logreg_proba'] = logreg.predict_proba(unknown_words_df)[:,1]

# Random Forest prediction dan probabilities
unknown['pred_forest'] = forest.predict(unknown_words_df)
unknown['pred_forest_proba'] = forest.predict_proba(unknown_words_df)[:,1]

# SVC prediction
unknown['pred_svc'] = svc.predict(unknown_words_df)

# Bayes prediction dan probabilities
unknown['pred_bayes'] = bayes.predict(unknown_words_df)
unknown['pred_bayes_probs'] = bayes.predict_proba(unknown_words_df)[:,1]

unknown
```
